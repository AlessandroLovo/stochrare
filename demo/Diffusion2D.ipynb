```python
import numpy as np
import matplotlib.pyplot as plt
from stochpy.dynamics.stochastic import Wiener, OrnsteinUhlenbeck
```

```python
ax = plt.axes()
ax.set_ylabel('x')
ax.set_ylabel('y')
ax.grid()
for n in xrange(4):
    t, x = Wiener().trajectory(np.array([0.,0.]),0.,dt=0.01)
    ax.plot(x[:,0],x[:,1])
plt.show()
```

```python
time, dist2 = zip(*[(t,r) for t,r in Wiener().sample_mean(np.array([0.,0.]),0.,1000,1000,dt=0.01,observable=lambda x: x[0]**2+x[1]**2)])
ax = plt.axes()
ax.set_xlabel('t')
ax.set_ylabel('r')
ax.grid()
ax.plot(time,np.sqrt(dist2))
plt.show()
```

```python
model = OrnsteinUhlenbeck(0,1,0.1)
xvec = np.linspace(-1.,1.)
yvec = np.linspace(-1.,1.)
potential = np.array([model.potential(np.array([x,y]),0) for x in xvec for y in yvec]).reshape(50,50)
ax = plt.axes()
ax.grid()
ax.axis('equal')
ax.contour(xvec,yvec,potential,20,colors='black')
t, x = model.trajectory(np.array([0.,0.]),0.,dt=0.01)
ax.plot(x[:,0],x[:,1])
plt.show()
```
