# Numerical solution of the Fokker-Planck equation

## Testing timestep

The time step *dt* should be much smaller than *dx^2/D*, with *dx* the spatial resolution and *D* the diffusivity

### Testing *dt* at fixed *dx=0.1* and varying *D*

```python
Wiener1D().pdfplot(1.0, 5.0, 10.0, dt=0.001, npts=400, bounds=(-20.0, 20.0), t0=0.0, th=True, P0='dirac', bc=('absorbing', 'absorbing'))
```

```python
Wiener1D(D=0.1).pdfplot(10.0, 100.0, dt=0.01, npts=400, bounds=(-20.0, 20.0), t0=0.0, th=True, P0='dirac', bc=('absorbing', 'absorbing'))
```

```python
Wiener1D(D=0.01).pdfplot(100.0, 1000.0, dt=0.1, npts=400, bounds=(-20.0, 20.0), t0=0.0, th=True, P0='dirac', bc=('absorbing', 'absorbing'))
```

```python
Wiener1D(D=10.0).pdfplot(0.1, 1.0, dt=0.0001, npts=400, bounds=(-20.0, 20.0), t0=0.0, th=True, P0='dirac', bc=('absorbing', 'absorbing'))
```

## Testing the numerical schemes

With the *implicit* and *Crank-Nicolson* schemes, we can afford using much larger timesteps than with the explicit method.

```python
Wiener1D().pdfplot(1.0, 10.0, dt=0.005, npts=400, bounds=(-20.0, 20.0), t0=0.0, th=True, P0='dirac', bc=('absorbing', 'absorbing'), method='explicit')
```
